```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('ano-2017.xlsx', engine='openpyxl', sheet_name='Ano-2017')
```

```python
variaveis = ['sgUF', 'sgPartido', 'datEmissao', 'vlrDocumento', 'vlrGlosa', 'vlrLiquido']
```

```python
data['datEmissao'] = pd.to_datetime(data['datEmissao'])
```

```python
data[variaveis].dtypes
```

```python
data['sgUF'].unique()
```

```python
data['sgPartido'].unique()
```

```python
todas_despesas = data.loc[data['vlrLiquido'].notnull()][variaveis]
todas_despesas.head(3)
```

```python
pt_despesas = data.loc[(data['sgPartido'] == 'PT') & (data['vlrLiquido'].notnull())][variaveis]
pt_despesas.head(3)
```

```python
pt_media_despesas = pt_despesas['vlrLiquido'].mean()
todas_media_despesas = todas_despesas['vlrLiquido'].mean()

print('Quantidade de transações do PT: ', pt_despesas.shape[0])
print('Valor médio das transações do PT: R$ ', round(pt_media_despesas, 2))
print('Quantidade de transações de todos: ', todas_despesas.shape[0])
print('Valor médio das transações de todos: R$ ', round(todas_media_despesas, 2))
```

```python
pt_acima_media_despesas = pt_despesas.loc[pt_despesas['vlrLiquido'] > pt_media_despesas]
todas_acima_media_despesas = todas_despesas.loc[todas_despesas['vlrLiquido'] > todas_media_despesas]

print('Quantidade de transações do PT acima do valor médio: ',
      pt_acima_media_despesas.shape[0])
print('Valor médio das transações do PT acima do valor médio: R$ ',
      round(pt_acima_media_despesas['vlrLiquido'].mean(), 2))
print('Quantidade de transações de todos acima do valor médio: ',
      todas_acima_media_despesas.shape[0])
print('Valor médio das transações de todos acima do valor médio: R$ ',
      round(todas_acima_media_despesas['vlrLiquido'].mean(), 2))
```

```python
psdb_rs_expenses = data.loc[(data['sgUF'] == 'RS') & (data['sgPartido'] == 'PSDB') & (data['vlrLiquido'].notnull())][variaveis]
psdb_rs_expenses.head(3)
```

```python
psdb_rs_mean_expenses = psdb_rs_expenses['vlrLiquido'].mean()

print('Quantidade de transações do PSDB RS: ', psdb_rs_expenses.shape[0])
print('Valor médio das transações do PSDB RS: R$ ', round(psdb_rs_mean_expenses, 2))
```

```python
recess_expenses = data.loc[(data['datEmissao'].dt.month.isin([1, 2, 12])) & (data['vlrLiquido'].notnull())][variaveis]

recess_expenses.head(3)
```

```python
recess_mean_expenses = recess_expenses['vlrLiquido'].mean()

print('Quantidade de transações do recesso: ', recess_expenses.shape[0])
print('Valor médio das transações do recesso: R$ ', round(recess_mean_expenses, 2))
```
