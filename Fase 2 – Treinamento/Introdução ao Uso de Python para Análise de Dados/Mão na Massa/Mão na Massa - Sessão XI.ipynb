d) Quais são os top 15 municípios com maior % de população até 14 anos? E com 60 ou mais? Quais os diferentes
insights extraídos?

```python
def transform_data(data, transform, result_column, columns, rows=15, ascending=False):
    result_series = data.apply(transform, axis=1).sort_values(ascending=ascending)
    result_data = data.loc[result_series.index][columns]
    result_data[result_column] = result_series.values
    return result_data.head(rows)
```

```python
# Quais são os top 15 municípios com maior % de população até 14 anos?

transform_data(
    base,
    lambda x: (x['POP_ABAIXO_1'] + x['POP_1_4'] + x['POP_5_9'] + x['POP_10-14']) / x['POP_ESTIMADA'] * 100,
    result_column='POP_ATE_14_%',
    columns=['CIDADE', 'ESTADO','POP_ESTIMADA']
)
```

```python
# Quais são os top 15 municípios com maior % de população com 60 anos ou mais?

transform_data(
    base,
    lambda x: x['POP_60_MAIS'] / x['POP_ESTIMADA'] * 100,
    result_column='POP_60_MAIS_%',
    columns=['CIDADE', 'ESTADO', 'POP_ESTIMADA']
)
```

e) Como é a distribuição de frequência dos municipios dentro de cada faixa de IDHM (<0.6, entre 0.6 e 0.65, 0.66 a
0.75 e >0.75). Como essa distribuição varia de uma região para outra do Brasil?


```python
# Primeiro criamos uma função que será aplicada como transformação da coluna IDHM para gerar um nova coluna de categorias.

def idhm_range(idhm):
    if pd.isnull(idhm):
        return 'Desconhecido'
    elif idhm < 0.6:
        return 'Ruim'
    elif idhm <= 0.65:
        return 'Razoável'
    elif idhm <= 0.75:
        return 'Bom' 
    else:
        return 'Elevado'
```

```python
# Em seguida criamos a função que será aplicada como transformação da coluna ESTADO para gerar uma nova coluna REGIAO.

def region(state):
    if state in ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']:
        return 'Norte'
    elif state in ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']:
        return 'Nordeste'
    elif state in ['DF', 'GO', 'MT', 'MS']:
        return 'Centro-Oeste'
    elif state in ['ES', 'MG', 'RJ', 'SP']:
        return 'Sudeste'
    elif state in ['PR', 'RS', 'SC']:
        return 'Sul'
    else:
        return 'Desconhecido'
```

```python
# Agora criamos uma cópia do dataset original e aplicamos as transformações.

idhm_by_region_data = base[['CIDADE', 'ESTADO', 'IDHM']].copy()
idhm_by_region_data['IDHM_CATEGORIA'] = base['IDHM'].apply(idhm_range)
idhm_by_region_data['REGIAO'] = base['ESTADO'].apply(region)
idhm_by_region_data.head()
```

```python
idhm_by_region_data = idhm_by_region_data.groupby(by=['REGIAO', 'IDHM_CATEGORIA']).aggregate({'CIDADE': 'count', 'IDHM': 'mean'})
idhm_by_region_data
```

```python
# Os mesmos resultados podem ser visualizados em outra perspectiva de colunas utilizando a função unstack.

idhm_by_region_data.unstack().round(2)
```
